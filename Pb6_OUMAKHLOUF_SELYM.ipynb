## 1. Relations binaires

Une **relation binaire** $\mathcal{R}$ sur un ensemble $E$ est un objet mathématique qui décrit une relation sur les couples d'éléments de $E$. Pour un couple $(x,y)$, il y a deux possibilités : ou bien $x$ est en relation avec $y$ (ce que l'on note $x\mathcal{R}y$), ou bien $x$ n'est pas en relation avec $y$ (ce que l'on note $x\cancel{\mathcal{R}}y$).

Soit $\mathcal{R}$ une relation binaire sur un ensemble $E$.
* On dit que la relation est **réflexive** si :

$$
\forall x \in E,\ x \mathcal{R} x.
$$

* On dit que la relation est **antiréflexive** si :

$$
\forall x \in E,\ x \cancel{\mathcal{R}} x.
$$

* On dit que la relation est **symétrique** si :

$$
\forall x \in E,\ \forall y \in E,\ x \mathcal{R} y \Rightarrow y \mathcal{R} x.
$$

* On dit que la relation est **transitive** si :

$$
\forall x \in E,\ \forall y \in E,\ \forall z \in E,\ (x \mathcal{R} y\ \text{et}\ y \mathcal{R} z) \Rightarrow x \mathcal{R} z.
$$


Toute relation binaire $\mathcal{R}$ sur un ensemble $E$ peut être définie en donnant l'ensemble des couples qui sont en relation. Quand l'ensemble $E$ est fini, il est possible de représenter (ou de définir) la relation binaire par une table d'adjacence ou un diagramme sagittal (un diagramme avec des flèches). 

Soit par exemple la relation binaire sur $\{0,1,2,3\}$ dont les couples d'éléments en relation sont $(0,2)$, $(1,2)$, $(1,3)$, $(2,2)$, $(2,3)$, $(3,0)$, $(3,1)$. La table d'adjacence et la représentation sagittale de cette relation sont figurées ci-dessous.

| 1 |   |   | x | x |
| 2 |   | x | x |   |
| 3 | x | x |   |   |



S'il est précisé que la relation binaire est symétrique, on peut la définir en ne donnant que les paires d'éléments en relation. Pour alléger le diagramme sagittal, on peut aussi se contenter de relier les éléments par un simple trait au lieu de deux flèches. 

Soit par exemple la relation binaire symétrique sur $\{0,1,2,3,4\}$ dont les paires d'éléments en relation sont $(0,0)$, $(2,1)$, $(2,2)$, $(3,0)$, $(3,1)$, $(4,0)$, $(4,1)$, $(4,3)$. **Théorème (de Ramsey).** Soit $p \geq 2$ et $q \geq 2$. Il existe un nombre $N$ tel que, dans tout groupe de $N$ personnes ou plus, il y a toujours :

## 3. Classe `RelationSymetrique`

* Créer une classe `RelationSymetrique` pour représenter et manipuler les relations binaires symétriques sur les ensembles du type $\{0,\ldots,n-1\}$, où $n\in\mathbb{N}^*$.
    * Un objet de cette classe aura pour attributs :
        * `n` : le nombre d'éléments de l'ensemble ;
        * `table`: la table d'adjacence (sous forme d'une liste de listes de booléens) décrivant la relation binaire.
    * La méthode `__init__` prendra en paramètre (en plus de `self`) le nombre d'éléments de l'ensemble et initialisera la table d'adjacence avec uniquement des `False` (cela revient à créer la relation binaire où aucun couple d'éléments n'est en relation).
    * Cette classe contiendra également les méthodes suivantes :
        * `met_relation` qui prend en paramètre deux éléments $i$ et $j$ et qui modifie la table d'adjacence de façon à ajouter la relation entre $i$ et $j$ (ainsi que la relation entre $j$ et $i$ puisqu'il s'agit d'une relation symétrique) ;
        * `met_liste_relations` qui prend en paramètre une liste de couples d'éléments et modifie la table d'adjacence de façon à ajouter les relations entre ces couples d'éléments ;
        * `supprime_relation` qui change la table d'adjacence pour de façon à **supprimer** la relation entre $i$ et $j$ (et entre $j$ et $i$) ;
        * `__str__` qui renvoie la liste des paires d'éléments en relation (sous forme d'une chaîne de charactères).
        
Pour compléter la description de cette classe, voici un exemple d'utilisation.    
~~~py
>>> R = RelationSymetrique(4)
>>> R.table
[[False, False, False, False],
 [False, False, False, False],
 [False, False, False, False],
 [False, False, False, False]]
>>> print(R)
[]

>>> R.met_relation(0,1)
>>> R.table
[[False, True, False, False],
 [True, False, False, False],
 [False, False, False, False],
 [False, False, False, False]]
 
>>> R.met_liste_relations([(0, 2), (0, 3), (1, 2), (2, 3)])
>>> R.supprime_relation(0, 3)
>>> print(R)
[(0, 1), (0, 2), (1, 2), (2, 3)]
~~~

```python
class RelationSymetrique :
    
    def __init__(self, n) :
        self.table = []
        self.n = n
        for _ in range(n) :
            self.table.append([False] * n)
            
            
    def met_relation(self, i, j):
        self.table[i][j] = True
        self.table[j][i] = True
        
    def met_liste_relations(self, indice):
        for _ in indice :
            self.met_relation(_[0], _[1])
            
    def supprime_relation(self, i, j):
        self.table[i][j] = False
        self.table[j][i] = False
        
    def __str__(self):
        couple = []
        for _ in range(len(self.table)) :
            for i in range(len(self.table)) :
                if self.table[_][i] == True  and (i, _) not in couple: # pour éviter les doublons de couples
                    couple.append((_, i))
                                  
        return str(couple)
    
    def suivant(self) :
        for _ in range(1, self.n) :
            for j in range(_) :   
                if not self.table[_][j] :    
                    self.met_relation(_, j)
                    return True
                self.supprime_relation(_, j)                   
        return False
```

* Modifier la fonction `affiche_diagramme` ci-dessous de façon à ce qu'elle affiche le diagramme d'une relation binaire symétrique.

```python
import matplotlib.pyplot as plt
from math import sin, cos, pi
from random import choice

def affiche_diagramme(R):
    
    n = R.n
    px, py = [], []
    
    choix_c = choice(["blue", "red", "black", "yellow", "green", "grey", "pink", "purple"])
    
    for i in range(n):
        px.append(cos(2 * pi * i / n))
        py.append(sin(2 * pi * i / n))
        
    for _ in range(n):
        for j in range(n):
            if _ <= j and R.table[_][j]:
                plt.plot([cos(2*pi*_/n), cos(2*pi*j/n)], [sin(2*pi*_/n), sin(2*pi*j/n)], color = choix_c)
                
    plt.axis('equal')
    plt.plot(px, py, linestyle="", marker="o", markersize=10.)
    plt.show()
    

R = RelationSymetrique(4)
R.met_relation(0, 1)
R.met_liste_relations([(0, 2), (0, 3), (1, 2), (2, 3)])
R.supprime_relation(0, 3)
affiche_diagramme(R)
```

## 4. Parcours de toutes les relations binaires symétriques antiréflexives

Le théorème de Ramsey ne fait jamais intervenir le fait qu'un élément soit en relation avec lui même ou non. On peut donc se limiter à l'exploration des relations binaires antiréflexives, ce qui réduit le nombre de cas.


Pour parcourir toutes les relations binaires symétriques antireflexives, on s'inspire d'un compteur binaire sur $k$ bits.

**Exemple avec $k=6$**. 
- On commence avec $k$ fois zero : `000000`.
- On obtient le nombre suivant en changeant le dernier $0$ en $1$: `000001`.
- Pour avoir celui d'après, c'est un peu plus compliqué car il y a une retenue quand on ajoute $1$ en binaire, il faut remettre le dernier bit à $0$ avant de mettre l'avant dernier à $1$:  `000010`.
- De façon générale, pour aller au suivant il faut, de droite à gauche, changer tous les $1$ en $0$ (propagation de la retenue), puis remplacer le $0$ le plus à droite par un $1$, s'il existe: `010111` devient `011000`.
- On s'arrête quand il n'y a que des $1$ : il n'y a pas de suivant.



- Ajouter à la classe `RelationSymetrique` une méthode `suivant` qui fonctionne ainsi.
    - Elle parcourt les couples $(i,j)$, avec $i<j$, dans l'ordre lexicographique : 
        - si $i \not R j$, alors on rajoute le fait que $i R j$ dans la relation (analogie avec le compteur : le 0 devient un 1) et on s'arrête et retourne `True`. On a changé la relation en la suivante.
        - si $i R j$, alors on enlève le fait que $i R j$ dans la relation (analogie avec le compteur : on propage la retenue et le 1 devient un 0) et on continue en passant au couple $(i,j)$ suivant.
    - Si on ne s'est pas arrêté au cours du parcours des couples $(i,j)$ c'est qu'on a fini de parcourir les relation binaires antireflexives, on s'arrête et retourne `False`.

## 5. Théorème de Ramsey pour $p=3$ et $q=3$

- Ecrire une fonction `groupe3amis` qui détermine s'il existe un groupe de 3 amis dans une relation `R`.

```python
def groupe3amis(R) :
    for _ in range(R.n):
        for j in range(_ + 1, R.n):
            for x in range(j + 1, R.n):
                if R.table[_][j] and R.table[_][x] and R.table[j][x]:
                    return True
    return False
```

- Ecrire une fonction `groupe3etrangers` qui détermine s'il existe un groupe de 3 étrangers dans une relation `R`.

```python
def groupe3etrangers(R):
    for _ in range(R.n):
        for i in range(_ + 1, R.n):
            for j in range(i + 1, R.n):
                if not (R.table[_][i] or R.table[_][j] or R.table[i][j]):
                    return True
    return False
```

- Ecrire une fonction `testRamsey33` qui prend en paramètre un entier `n` et teste si toutes les relations binaire symétriques antiréflexives sur $\{0,\ldots,n-1\}$ vérifient la propriété de Ramsey pour $p=3$ et $q=3$.

```python
def testRamsey33(n):
    R = RelationSymetrique(n)
    for _ in range(2 ** ((n * (n - 1) // 2))):
        if not (groupe3amis(R) or groupe3etrangers(R)):
            return False
        R.suivant()
    return True
```

## 7. Théorème de Ramsey pour $p=4$ et $q=3$

- Ecrire une fonction `groupe4Amis` qui détermine s'il existe un groupe de 4 amis dans une relation `R`.

```python
def groupe4amis(R) :
    for _ in range(R.n):
        for j in range(_ + 1, R.n):
            for x in range(j + 1, R.n):
                for k in range(x + 1, R.n):
                    if R.table[_][j] and R.table[_][x] and R.table[_][k] and R.table[j][x] and R.table[j][k] and R.table[x][k]:
                        return True
    return False
```

- Ecrire une fonction `ctrexRamsey43` qui prend en paramètre un entier `n` et renvoie une relation binaire symétrique sur $\{0,\ldots,n-1\}$ qui ne vérifie pas la propriété de Ramsey pour $p=4$ et $q=3$ (ou renvoie `None` s'il n'en existe pas).

```python
def ctrexRamsey43(n):
    R = RelationSymetrique(n)
    for _ in range (2 ** ((n * (n - 1) // 2))):
        if not (groupe4amis(R) or groupe3etrangers(R)):
            return R
        R.suivant()
```
